## Data pre-processing

## Data Analysis 

### Gene expression analysis using means

1) Για κάθε γονίδιο βρειτε τη μέση τιμή ανα κλάση (Cancer, Normal) και τη διαφορά τους.

```python
sorted_genes['CancerMean'] = sorted_genes.iloc[:,1:11].mean(axis=1)
sorted_genes
```

```python
sorted_genes['NormalMean'] = sorted_genes.iloc[:, 11:21].mean(axis=1)
sorted_genes
```

```python
sorted_genes['DiffMeans'] = sorted_genes['CancerMean'] - sorted_genes['NormalMean']
sorted_genes # genes with all means
```

2) Ταξινομείστε τα γονιδια με βάση τη διαφορά στους μέσους όρους

```python
sorted_genes1 = sorted_genes.sort_values(by=['DiffMeans'])
sorted_genes1 # sorted genes by diffmeans
```

3) δημιουργείστε ένα heatmap για 100 γονίδια όπου τα 50 πρώτα θα είναι τα γονίδια με τη μέγιστη τιμή στη διαφορά και τα αλλα 50 γονίδια με την ελάχιστη τιμή στη διαφορά


```python
sns.heatmap(sorted_genes1.iloc[np.r_[0:50, -50:0], 1:-3])
```

### Gene expression analysis using p-value

1) Για κάθε γονίδιο βρείτε το p-value (μπορείτε να χρησιμοποιήσετε την συνάρτηση ttest_ind).


```python
import scipy.stats as stats
```

```python
pvalue_genes = sorted_genes.copy()
pvalList = []
for x in range(pvalue_genes.shape[0]):
    st, pval = stats.ttest_ind( pvalue_genes.iloc[x, 1:11] ,pvalue_genes.iloc[x, 11:21] )
    pvalList.append(pval)
pvalue_genes['PValue'] = pvalList; # add Pvalues
pvalue_genes #sorted genes with means and P value
```

2) Ταξινομείστε τα γονίδια με βάση τη διαφορά στους μέσους όρους.

```python
sorted_pvalue_genes = pvalue_genes.sort_values(by=['DiffMeans'])
sorted_pvalue_genes
```

3) Δημιουργείστε ένα heatmap για 100 γονίδια. Τα 50 πρώτα θα είναι τα γονίδια με τη μέγιστη τιμή στη διαφορά των μέσων όρων και τιμή p-value κάτω από 0.0001. Τα άλλα 50 γονίδια με την ελάχιστη τιμή στη διαφορά των μέσων όρων και τιμή p-value κάτω από 0.0001.


```python
spv_P = sorted_pvalue_genes[(sorted_pvalue_genes['PValue'] < 0.0001)]
spv_P # genes where Pvalue < 0.0001
```

```python
sns.heatmap(spv_P.iloc[np.r_[0:50, -50:0], 1:-4])
```

### Gene expression analysis using q-value

1)  Για κάθε γονίδιο βρείτε το q-value με βάση το p-value. 

```python
qvalue_genes = pvalue_genes.copy()
qvalue_genes #sorted genes with means , P and Q value
```

```python
qvalList = []
n = len(qvalue_genes)
for x in range(qvalue_genes.shape[0]):
    qval = qvalue_genes['PValue'][x] * (n / (n - x))
    qvalList.append(qval) 
```

```python
qvalue_genes['QValue'] = qvalList.copy()
qvalue_genes
```

2) Ταξινομείστε τα γονίδια με βάση τη διαφορά στους μέσους όρους

```python
sorted_qvalue_genes = qvalue_genes.sort_values(by=['DiffMeans'])
sorted_qvalue_genes
```

3) Δημιουργείστε ένα heatmap όλα για 100 γονίδια .Τα 50 πρώτα θα είναι τα γονίδια με τη μέγιστη τιμή στη διαφορά των μέσων όρων και τιμή q-value <=0.05. Τα άλλα 50 γονίδια με την ελάχιστη τιμή στη διαφορά των μέσων όρων και τιμή q-value <=0.05.

```python
sqv_Q = sorted_qvalue_genes[(sorted_qvalue_genes['QValue'] <= 0.05)]
sqv_Q # genes where Qvalue <= 0.05
```

```python
sns.heatmap(sqv_Q.iloc[np.r_[0:50, -50:0], 1:-5])
```

ii) Σε ποια hallmark gene list (πρώτη κολώνα στα δεδομένα του
h.all.v7.1.entrez.gmt αρχείου) έχετε τα περισσότερα γονίδια.

```python
# A list of sets(every set includes values of every row ,
# in hallmark_genes)
hallmark_genes_list =[] 
mini_set = set()
for x in range(len(hallmark_genes.index)):
    for y in range (len(hallmark_genes.columns)):
        if(pd.notnull(hallmark_genes.iloc[x , y])):
            mini_set.add(hallmark_genes.iloc[x , y])
    hallmark_genes_list.append(mini_set.copy())
    mini_set.clear()
```

```python
cnt_means = {}  # Dictionaries that each element is the number of row and the total count of values equals to hallmark genes 
cnt_pvalue = {}  
cnt_qvalue = {}
mcnt = 0
pcnt = 0
qcnt = 0
cnt = 0
for x in hallmark_genes_list: # for means
    for y in (means_genes_set):
        if ( y in x ):
            mcnt+=1
    cnt_means.update({cnt:mcnt})
    mcnt = 0
    cnt += 1

cnt = 0

for x in hallmark_genes_list: # for pvalues
    for y in (pvalue_genes_set):
        if ( y in x ):
            pcnt += 1
    cnt_pvalue.update({cnt:pcnt})
    pcnt = 0
    cnt += 1

cnt = 0

for x in hallmark_genes_list: # for qvalues
    for y in (qvalue_genes_set):
        if ( y in x ):
            qcnt+=1
    cnt_qvalue.update({cnt:qcnt})
    qcnt = 0
    cnt += 1
```

```python
max_means = {}
max_pvalues = {}
max_qvalues = {}
max = -1
max_thesis = -1
for x in cnt_means:     # Means
    if(cnt_means.get(x) >= max):
        max = cnt_means.get(x)
        max_thesis = x
        max_means.update({max_thesis : max})
for y in max_means.copy():
    if( max_means.get(y) < max):
        max_means.pop(y)
        
max = -1                
max_thesis = -1

for x in cnt_pvalue:       # Pvalue 
    if(cnt_pvalue.get(x) >= max):
        max = cnt_pvalue.get(x)
        max_thesis = x
        max_pvalues.update({max_thesis : max})
for y in max_pvalues.copy():
    if( max_pvalues.get(y) < max):
        max_pvalues.pop(y)
        
max = -1                
max_thesis = -1
for x in cnt_qvalue:       # Qvalue 
    if(cnt_qvalue.get(x) >= max):
        max = cnt_qvalue.get(x)
        max_thesis = x
        max_qvalues.update({max_thesis : max})
for y in max_qvalues.copy():
    if( max_qvalues.get(y) < max):
        max_qvalues.pop(y)
```

```python
max_means_list = max_means.keys()  # store number of line where max is found
max_qvalue_list = max_qvalues.keys()
max_pvalue_list = max_pvalues.keys()
```

```python
print("Most genes selected by means are in list :")
for x in max_means_list:
    print (hallmark_genes.index[x])

print("\nMost genes selected by pvalue are in list :")
for x in max_pvalue_list:
    print (hallmark_genes.index[x])

print("\nMost genes selected by qvalue are in list :")
for x in max_qvalue_list:
    print (hallmark_genes.index[x])
```
